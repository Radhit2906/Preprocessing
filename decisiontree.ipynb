```python
#Import library numpy, pandas dan scikit-learn
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
#Membaca Dataset dari File ke Pandas dataFrame
CobaDataset = pd.read_csv('DatasetCoba1.csv', delimiter=';', header=0)
```

```python
CobaDataset.head()
```

```python
# Mengubah kelas kolom (penerima sembako)label menjadi unique integer
CobaDataset["SEMBAKO"] = pd.factorize(CobaDataset.SEMBAKO)[0]
```

```python
# Mengubah kelas kolom (penerima sembako)label menjadi unique integer
CobaDataset["Pekerjaan"] = pd.factorize(CobaDataset.Pekerjaan)[0]
CobaDataset["Prioritas"] = pd.factorize(CobaDataset.Prioritas)[0]
CobaDataset["JenisKelamin"] = pd.factorize(CobaDataset.JenisKelamin)[0]
CobaDataset["HubunganKepalaKeluarga"] = pd.factorize(CobaDataset.HubunganKepalaKeluarga)[0]
```

```python
CobaDataset.head()
```

```python
print(CobaDataset)
```

```python
CobaDataset = CobaDataset.drop(labels="No", axis=1)
```

```python
CobaDataset = CobaDataset.drop(labels="Kecamatan", axis=1)
```

```python
print(CobaDataset)
```

```python
CobaDataset = CobaDataset.to_numpy()
```

```python
print(CobaDataset)
```

```python
#Membagi Dataset => 80 baris data untuk training dan 20 baris data untuk testing
dataTraining = np.concatenate((CobaDataset[0:100, :], CobaDataset[100:1000, :]),
                              axis=0)
dataTesting = np.concatenate((CobaDataset[1000:2000, :], CobaDataset[2000:15000, :]),
                              axis=0)
```

```python
print(dataTraining)
len(dataTraining)
```

```python
# memecah dataset ke input dan label
inputTraining = dataTraining[:, 0:6]
inputTesting = dataTesting[:, 0:6]
labelTraining = dataTraining[:, 6]
labelTesting = dataTesting[:, 6]

print(labelTraining)
len(labelTraining)
```

```python
model = tree.DecisionTreeClassifier()
```

```python
# Training Model
model = model.fit(inputTraining, labelTraining)
```

```python
#Memprediksi Input Data Testing
hasilPrediksi = model.predict(inputTesting)
print("Label Sebenarnya : ", labelTesting)
print("Hasil Prediksi : ", hasilPrediksi)
```

```python
#Menghitung Akurasi
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar :", prediksiBenar, "data")
print("Prediksi Salah :", prediksiSalah, "data")
print("Akurasi :", prediksiBenar/(prediksiBenar+prediksiSalah) * 100, "%")
```
